 **Filter the titles which contain python**

```python
df_title =df[df["title"].str.contains("Python" or "python" or "PYTHON")]
print(df_title["title"])

# Another method
def title_with_python():
    n = 1
    for title in df.title:
        if ("Python" or "python" or "PYTHON") in title:  
            print(n, title)
            n +=1
    return " "
twp = title_with_python()
print(twp)
```

**Filter the titles which contain JavaScript**

```python
df_title =df[df["title"].str.contains("JavaScript" or "javascript" or "JAVASCRIPT")]
print(df_title["title"])

# Another method
def title_with_JavaScript():
    n = 1
    for title in df.title:
        if ("JavaScript" or "javascript" or "JAVASCRIPT") in title:  
            print(n, title)
            n +=1
    return " "
twj = title_with_JavaScript()
print(twj)
```
